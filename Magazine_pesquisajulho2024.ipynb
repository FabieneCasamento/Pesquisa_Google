```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
# criar o navegador
navegar = webdriver.Chrome()

navegar.get('https://www.magazineluiza.com.br/')


print('Desejo pesquisar sobre notebook')
produto= "Notebook"

```

```python
#apertar o enter, igual fazemos no site normal
#copiar o xpath
navegar.find_element('xpath', '//*[@id="input-search"]').send_keys(produto,Keys.ENTER)
```

```python
#inspecionar os dados no site com a janela abaixo dos dados
#lista_resultados_magazine = navegar.find_elements('class name', 'sc-dcjTxL xDJfk')
#sempre coletar os ultimos dados da lista
#lista_resultados_magazine = navegar.find_elements('class name', 'xDJfk')

#testid="productCard-shipping-tag" class="sc-fqkvVR keRZJx"></div></div>
lista_resultados_magazine = navegar.find_elements('class name', 'fBmkOc')



```

```python
#pegar os resultados
import time
time.sleep(10)
coletar_dados=[]
for resultado in lista_resultados_magazine:
    nome= resultado.find_element(By.TAG_NAME,'h2').text
    print(nome)
    #colocar sempre a última parte depois do espaco
    preco= resultado.find_element('class name', 'dOwMgM').text
    print(preco)
    preco_2= preco.replace('R$','').replace('','').replace('.','').replace(',','.')
    preco_2= float(preco_2)
    print(preco_2)
    registros= [nome,preco_2]
    coletar_dados.append(registros)
```
