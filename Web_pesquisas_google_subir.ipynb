```python
import pathlib
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email.mime.text import MIMEText  # Importando o MIMEText
from email import encoders


caminho_arquivo = pathlib.Path(r'')

#enviar o e-mail *(preparando o terreno)
email_enviar_mail_To = 'MEUEMAIL+pesquisa@gmail.com'
titulo_email_Subject = f'Produto(s) Encontrado(s) na faixa de preço desejada do {produto}'
#salvo caminho
#attachmet
#tem que ter espaço entre /
attachmet_archive = pathlib.Path.cwd() / f'{arquivo_pesquisa_escrito_excel}'

print("Caminho do arquivo:", attachmet_archive)    

attachmet_archive_complete = pathlib.Path.cwd() / f'{Toda_pesquisa_escrito_excel}'



def enviar_email(CORPO_HTMLBody):
    # Criar a mensagem de e-mail
    msg = MIMEMultipart()
    msg['Subject'] = titulo_email_Subject
    #from meu e-mail
    msg['From'] = 'MEUEMAIL@gmail.com' 
    #e-mail a ser enviado , para teste seu email + nomequalquer @gmail.com
    msg['To'] = email_enviar_mail_To
    #sua senha de duas etapas do email gmail
    senhago ='MINHA SENHA DA CHAVE DE E-MAIL'
    password = senhago
    # Corpo do e-mail
    #msg.attach(MIMEText(CORPO_HTMLBody, 'plain'))
    #escrever de forma html, o acima é texto apenas
    msg.attach(MIMEText(CORPO_HTMLBody, 'html'))

    #excel
    # Anexar o arquivo
    with open(attachmet_archive, 'rb') as arquivo:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload((arquivo).read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', f"attachment; filename= {arquivo_pesquisa_escrito_excel}",)
        msg.attach(part)   

    #arquivo completo
    with open(attachmet_archive_complete, 'rb') as arquivo:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload((arquivo).read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', f"attachment; filename= {Toda_pesquisa_escrito_excel}",)
        msg.attach(part) 
        
    
    

    # Conectar-se ao servidor SMTP e enviar o e-mail
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(msg['From'], password)
    server.send_message(msg)
    server.quit()

    print('Email enviado completo')


#verificando se existe alguma oferta dentro da tabela de ofertas
if len(preco_ordem_decrescente.index) > 0:
    print('Pode enviar por e-mail')

    Texto_HTMLBody = f"""
    <p>Prezados,</p>
    <p>Encontramos alguns produtos em oferta dentro da faixa de preço desejada. Segue tabela com detalhes</p>
    <br>
    <p>Onde a pesquisa por preço desejado foi de <strong>R$ {preco_minimo:.0f}</strong> a <b>R$ {preco_maximo:.0f}</b></p>
    {preco_ordem_decrescente_curto.to_html(index=False)}
    <p>Qualquer dúvida estou à disposição</p>
    <p>Att.,</p>"""
    enviar_email(Texto_HTMLBody)

```
