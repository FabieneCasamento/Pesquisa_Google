# Pesquisa de cotação de dinheiro rápido, pelo google

```python
#!pip install selenium
#criar um navegador
#pesquiso no google => selenium with python
#https://selenium-python.readthedocs.io/getting-started.html
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
# criar o navegador
navegar = webdriver.Chrome()

navegar.get('https://www.google.com/')

money_search = "Cotação do Dólar"
print(f'Desejo pesquisar sobre {money_search}')



import time
#funções

def coleta_cotacao(money_search_funcao, codigo_class, escrita_real):

    navegar.get('https://www.google.com/')
    print(f'Desejo pesquisar sobre {money_search_funcao}')
    # colocar o nome da pesquisa e apertar o enter, igual fazemos no site normal
    navegar.find_element('xpath', '//*[@id="APjFqb"]').send_keys(money_search_funcao,Keys.ENTER)
    # Encontrar o elemento que contém a cotação do dólar
    time.sleep(3)
    valor_reais = navegar.find_elements(By.CLASS_NAME, codigo_class)
    print(valor_reais)
    # Para visualizar o texto do elemento encontrado
    if money_search_funcao in 'Cotação do Ouro em reais':
        numero=0
        for valor in valor_reais:
            print(valor.text)
            valor_real = valor.text
            numero =numero +1
            if numero == 1: 
               valor_real_final = valor.text 
        valor_real =valor_real_final
    else:
        for valor in valor_reais:
            print(valor.text)
            valor_real = valor.text
                
    print(valor_real)    
    valor_real = valor_real.replace('O valor do grama do ouro hoje está cotado em R$ ','')
    print(valor_real)
    valor_real = valor_real.replace('.', '').replace(',', '.')
    print('final')
    print(valor_real)
    valor_real =float(valor_real)
    # Encontrar o elemento que contém a cotação do dólar
    escrito_dolar_reais = navegar.find_elements(By.CLASS_NAME, escrita_real)
    print(escrito_dolar_reais)
    escrito_dolar_reais2 = escrito_dolar_reais[0].text 
    return money_search_funcao, valor_real, escrito_dolar_reais2

```

```python
import pathlib
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email.mime.text import MIMEText  # Importando o MIMEText
from email import encoders

def enviar_email_FONTEGMAIL(texto_hora_1):
    path_excel= 'cotacao.xlsx'
    arquivo_escrito_noemail = 'cotacao_new.xlsx'

    #enviar o e-mail *(preparando o terreno)
    email_enviar_mail_To = Cod_key_email.teste_envio
    titulo_email_Subject = f'Cotação atual das moedas e do Ouro'
    
    #salvo caminho
    #attachmet
    #tem que ter espaço entre /
    attachmet = path_excel
    print("Caminho do arquivo:", attachmet)       



    #imprimir
    print('*'*50)
    print('dados base para envio em html')
    # COMMAND  tabela ----------
   
    def highlight_max(s):
      is_max = s == s.max()
      return ['background-color: blue' if v else '' for v in is_max]   
       
    #styled_df = tabela.style.apply(highlight_max, axis=0)
    
    # Aplicar a estilização e formatar a coluna
    styled_df = tabela.style.apply(highlight_max, axis=0).format({'Cotação em Reais': '{:.2f}'})

    print('style')
    #styled_df['Cotação em Reais'] = styled_df['Cotação em Reais'].astype('float')
    #styled_df['Cotação em Reais']= styled_df['Cotação em Reais'].round(2)
    
    
    # Exibir a tabela estilizada
    display(styled_df)
    
    
    # COMMAND ----------
    print('style html')
    print(styled_df.to_html())
    
    # COMMAND ----------
    
    table_styled_df= styled_df.to_html()
       
    #------------------------------------------------

    body_email= f"""
    <!DOCTYPE html>
    <html>
    <head>
    <center>
    <title>Arquivo com cotação geral das moedas</title>
    </center>
    </head>
    <body>
    
    
    
    <center>
    <h1>Gráfico do resultado da busca de cotações de moedas:</h1>
    
    <h1>Pesquisa geral da cotação. Além disso, {texto_hora_1} </h1>
    Em anexo a tabela.
    </center>
    
    <center>{table_styled_df} </center>
    </style>
    </body>
    </html>
    """
    #----------------------corpo do e-mail----------------------
    print('corpo do e-mail')
    print(body_email)

    #----------------------corpo do e-mail----------------------
    
    #---------------------para enviar realmente o e-mail automático---------------------     
    # Criar a mensagem de e-mail
    msg = MIMEMultipart()
    msg['Subject'] = titulo_email_Subject
    #from meu e-mail
    msg['From'] = Cod_key_email.e_mail_meu
    #e-mail a ser enviado , para teste seu email + nomequalquer @gmail.com
    msg['To'] = email_enviar_mail_To
    #sua senha de duas etapas do email gmail
    password = Cod_key_email.chavekey
    # Corpo do e-mail
    #msg.attach(MIMEText(CORPO_HTMLBody, 'plain'))
    #escrever de forma html, o acima é texto apenas => body_email
    msg.attach(MIMEText(body_email, 'html'))

    #excel
    # Anexar o arquivo
    with open(attachmet, 'rb') as arquivo:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload((arquivo).read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', f"attachment; filename= {arquivo_escrito_noemail}",)
        msg.attach(part)
        

    # Conectar-se ao servidor SMTP e enviar o e-mail
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(msg['From'], password)
    server.send_message(msg)
    server.quit()

    print('Email enviado')




    


```

```python
import pathlib
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email.mime.text import MIMEText  # Importando o MIMEText
from email import encoders

def enviar_email_FOTO(body_email_2):
    path_excel= 'cotacao.xlsx'
    arquivo_escrito_noemail = 'cotacao_new.xlsx'

    #enviar o e-mail *(preparando o terreno)
    email_enviar_mail_To = Cod_key_email.teste_envio
    titulo_email_Subject = f'Cotação atual das moedas e do Ouro Completo'
    
    #salvo caminho
    #attachmet
    #tem que ter espaço entre /
    attachmet = path_excel
    print("Caminho do arquivo:", attachmet)       


    
    #---------------------para enviar realmente o e-mail automático---------------------     
    # Criar a mensagem de e-mail
    msg = MIMEMultipart()
    msg['Subject'] = titulo_email_Subject
    #from meu e-mail
    msg['From'] = Cod_key_email.e_mail_meu
    #e-mail a ser enviado , para teste seu email + nomequalquer @gmail.com
    msg['To'] = email_enviar_mail_To
    #sua senha de duas etapas do email gmail
    password = Cod_key_email.chavekey
    # Corpo do e-mail
    #msg.attach(MIMEText(CORPO_HTMLBody, 'plain'))
    #escrever de forma html, o acima é texto apenas => body_email
    msg.attach(MIMEText(body_email_2, 'html'))

    #excel
    # Anexar o arquivo
    with open(attachmet, 'rb') as arquivo:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload((arquivo).read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', f"attachment; filename= {arquivo_escrito_noemail}",)
        msg.attach(part)
        
    attachmet_figura = f'{nome_figura}.png'
    figura_nome = f'{nome_figura}_new.png'
    
    with open(attachmet_figura, 'rb') as arquivo_fig:
        part = MIMEBase('application', 'octet-stream')
        part.set_payload((arquivo_fig).read())
        encoders.encode_base64(part)
        part.add_header('Content-Disposition', f"attachment; filename= {figura_nome}",)
        msg.attach(part)
        
    
    # Conectar-se ao servidor SMTP e enviar o e-mail
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login(msg['From'], password)
    server.send_message(msg)
    server.quit()

    print('Email enviado')



```

```python
import tkinter as tk


def enviar():
    print(var_enviar_ou_nao.get())
    if var_enviar_ou_nao.get() in "Enviar por e-mail a cotação de hoje":
       print('enviar')
       enviar_email_FOTO(body_email)


janela = tk.Tk()

#variável string
var_enviar_ou_nao= tk.StringVar(value="Nada marcado")


botao_enviar_agora = tk.Radiobutton(text="Enviar por e-mail a cotação de hoje", variable=var_enviar_ou_nao, value="Enviar por e-mail a cotação de hoje")
botao_NAOenviar_agora = tk.Radiobutton(text="Não enviar a cotação por e-mail", variable=var_enviar_ou_nao, value="Não enviar a cotação por e-mail")
botao_enviar_agora.grid(row=0, column=0)
botao_NAOenviar_agora.grid(row=0, column=1)


#quero o botão para funcionar
botao_enviar = tk.Button(text='Enviar', command=enviar)
botao_enviar.grid(row=1, column=0)

janela.mainloop()
```
