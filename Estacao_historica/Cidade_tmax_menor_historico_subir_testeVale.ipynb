```python
print('Dados das menores temperaturas mínimas da história na cidade de {0} em um mês de {1}, dados desde {2}'.format('Campinas', mes_escrito,ano_escrito))
print('Fonte:Ciiagro')
print('=' * 40)
print('|{:^3}°)|{:^15}|{:^5}|{:^5}|{:^5}|'.format('Pos.','T. Min(°C)', 'Data','Mês', 'Ano'))
print('=' * 40)



conta = 0
for linha in Menor_TMin_novembro.index:
     #nome = tabela_produtos.loc[linha, coluna]
    Tmin = Menor_TMin_novembro.loc[linha, "tmin"]
    dia1 = int(Menor_TMin_novembro.loc[linha, "Dia_tabela"])   
    mes1 = int(Menor_TMin_novembro.loc[linha, "Mes_tabela"])
    ano1 = int(Menor_TMin_novembro.loc[linha,'Ano_tabela'] )
    Tmin2 = float(Tmin)
    temperatura1 = round(Tmin2,1)   
    
   # print(temperatura1)
    
    if temperatura1 < 5:
        print('|{:^3}°)|\033[42m{:^15}\033[0m|{:^5}|{:^5}|{:^5}|'.format(conta+1,temperatura1,dia1, mes1,ano1 ))

    else:
        print('|{:^3}°)|{:^15}|{:^5}|{:^5}|{:^5}|'.format(conta+1,temperatura1,dia1, mes1,ano1 ))
        
        pass
    conta = conta +1

print('=' * 40)
```

```python
print('Dados das MAIORES temperaturas máximas da história na cidade de {0} em um mês de {1}, dados desde {2}'.format('Campinas', mes_escrito, ano_escrito))
print('Fonte:Ciiagro')
print('=' * 40)
print('|{:^3}°)|{:^15}|{:^5}|{:^5}|{:^5}|'.format('Pos.','T. Máx(°C)', 'Data','Mês', 'Ano'))
print('=' * 40)



conta = 0
for linha in Menor_TMax_novembro.index:
     #nome = tabela_produtos.loc[linha, coluna]
    Tmin = Menor_TMax_novembro.loc[linha, "tmax"]
    dia1 = int(Menor_TMax_novembro.loc[linha, "Dia_tabela"])   
    mes1 = int(Menor_TMax_novembro.loc[linha, "Mes_tabela"])
    ano1 = int(Menor_TMax_novembro.loc[linha,'Ano_tabela'] )
    Tmin2 = float(Tmin)
    temperatura1 = round(Tmin2,1)   
    
   # print(temperatura1)
    
    if temperatura1 > 30:
        print('|{:^3}°)|\033[45m{:^15}\033[0m|{:^5}|{:^5}|{:^5}|'.format(conta+1,temperatura1,dia1-1, mes1,ano1 ))

    else:
        print('|{:^3}°)|{:^15}|{:^5}|{:^5}|{:^5}|'.format(conta+1,temperatura1,dia1-1, mes1,ano1 ))
        
        pass
    conta = conta +1

print('=' * 40)
```
